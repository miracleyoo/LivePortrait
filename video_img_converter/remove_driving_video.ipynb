```python
import os
import os.path as osp
import glob
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle as pkl
```

```python
augmented_video_root = '/prj/qct/mmrd-cv/esper/Misc0002/dataset/liveportrait-augmentation/augmented_videos'
video_list = glob.glob(osp.join(augmented_video_root, 'ning*', '*concat.mp4'))
print(len(video_list))
```

```python
# Load each video from video list, crop the first 1/3 width, and only output the rest 2/3 to it original video path
for video_path in video_list:
    print(video_path, '-->')
    subdir, video_name = osp.split(video_path)
    video_name = osp.splitext(video_name)[0]
    # video_path = osp.join(augmented_video_root, video_name)
    video_cap = cv2.VideoCapture(video_path)
    fps = video_cap.get(cv2.CAP_PROP_FPS)
    width = int(video_cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video_cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    frame_count = int(video_cap.get(cv2.CAP_PROP_FRAME_COUNT))
    # print(fps, width, height, frame_count)
    
    # Crop the first 1/3 widthv2.CAP_PROP_POS_FRAMES, int(fps * 1)
    out_path = osp.join(subdir, video_name + '_crop.mp4')
    print(out_path)
    out_video = cv2.VideoWriter(out_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width//3*2, height))
    for i in range(int(fps * 1), frame_count):
        ret, frame = video_cap.read()
        if not ret:
            break
        # print(width, height, frame[:, width//3:, :].shape)
        out_video.write(frame[:, width//3:, :])
    video_cap.release()
    out_video.release()
    # break
```

```python
# Remove all original videos that end with 'concat.mp4'
for video_path in video_list:
    if video_path.endswith('concat.mp4'):
        os.remove(video_path)
    
# Rename all processed video that end with 'concat_crop.mp4' to 'concat.mp4'
for video_path in video_list:
    if video_path.endswith('concat_crop.mp4'):
        os.rename(video_path, video_path.replace('concat_crop.mp4', 'concat.mp4'))
```
